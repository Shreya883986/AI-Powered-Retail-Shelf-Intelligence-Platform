# Import Libraries

```python
import pandas as pd
import numpy as np
```

# Load Datasets

```python
sales = pd.read_csv("../datasets/data warehouse/dw_fact_sales.csv")

products = pd.read_csv("../datasets/data warehouse/dw_dim_product.csv")

stores = pd.read_csv("../datasets/data warehouse/dw_dim_store.csv")

shelf = pd.read_csv("../datasets/data warehouse/dw_fact_shelf_audit.csv")
```

# Check Missing Values

```python
print("Sales")
print(sales.isnull().sum())

print("\nProducts")
print(products.isnull().sum())

print("\nStores")
print(stores.isnull().sum())

print("\nShelf Audit")
print(shelf.isnull().sum())
```

# Remove Duplicate Records

```python
sales = sales.drop_duplicates()

products = products.drop_duplicates()

stores = stores.drop_duplicates()

shelf = shelf.drop_duplicates()

print("Duplicates Removed")
```

# Check Data Types

```python
sales.dtypes
```

```python
products.dtypes
```

```python
stores.dtypes
```

```python
shelf.dtypes
```

# Convert Dates

```python
sales["transaction_date"] = pd.to_datetime(
    sales["transaction_date"]
)

shelf["audit_date"] = pd.to_datetime(
    shelf["audit_date"]
)
```

# Verify Date Conversion

```python
sales.info()
```

```python
shelf.info()
```

# Handle Missing Values

```python
sales = sales.fillna(0)

products = products.fillna("Unknown")

stores = stores.fillna("Unknown")

shelf = shelf.fillna("Unknown")
```

```python
print(sales.isnull().sum().sum())
print(products.isnull().sum().sum())
print(stores.isnull().sum().sum())
print(shelf.isnull().sum().sum())
```

# Save Cleaned Files

```python
sales.to_csv(
    "../datasets/processed datasets/clean_sales.csv",
    index=False
)

products.to_csv(
    "../datasets/processed datasets/clean_products.csv",
    index=False
)

stores.to_csv(
    "../datasets/processed datasets/clean_stores.csv",
    index=False
)

shelf.to_csv(
    "../datasets/processed datasets/clean_shelf_audit.csv",
    index=False
)
```
